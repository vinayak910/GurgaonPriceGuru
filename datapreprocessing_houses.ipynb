```python
import re
```

```python
df.society = df.society.apply(lambda name : re.sub(r'\d+(\.\d+)?\s?★','' , str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df.society = df['society'].str.replace('nan','independent')
```

```python
df.price.str.split().str[1].value_counts()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def treat_price(x):
    if type(x) == float:
        return x
    else:
        if x[1] == 'Lac':
            return round(float(x[0])/100,2)
        else:
            return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(treat_price)
```

```python
# price_per_sqft
df[~df['price_per_sqft'].str.contains('sq.ft.',na = False)]
```

```python
df.price_per_sqft.head()
```

```python
df.price_per_sqft = df.price_per_sqft.str.split('/').str[0].str.split('₹ ').str[1].str.replace(',','').str.strip().astype('float')
```
