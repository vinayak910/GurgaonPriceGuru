```python
fig, axs = plt.subplots(1, 2, figsize=(12, 6))  # 1 row, 2 columns

# Pie chart for 'flat' property type
flat_counts = df[df['property_type'] == 'flat']['agePossession'].value_counts(normalize=True)
axs[0].pie(flat_counts, labels=flat_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Blues"))
axs[0].set_title('Age of Possession for Flats')

# Pie chart for 'house' property type
house_counts = df[df['property_type'] == 'house']['agePossession'].value_counts(normalize=True)
axs[1].pie(house_counts, labels=house_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Greens"))
axs[1].set_title('Age of Possession for Houses')

# Display the plots
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
